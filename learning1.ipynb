```python
import seaborn as sns
import matplotlib.pyplot as plt

ship = sns.load_dataset("titanic")
print(ship)
sns.lineplot(x="sex", y="fare", data=ship)


plt.grid(True)
plt.title("Ship Data")
plt.xlabel("Sex")
plt.ylabel("Fare")
```

```python
sns.barplot(x="sex", y="fare", data=ship, linewidth=2.5, facecolor=(1,.3,.3,.5), errcolor="0.5", edgecolor=".2")
```
